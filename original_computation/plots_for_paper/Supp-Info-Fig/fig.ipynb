```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('renewal.times.txt', 'r') as file:
    lines = file.readlines()
```

```python
values = []
for x in lines[:]:
    x = x.strip()
    x = x.split()
    for x in x:
        # print(float(x))
        values.append(float(x))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
ax.hist(values, bins=15, histtype='bar', edgecolor='black', linewidth=1.5, align='mid', density=False, rwidth=5, alpha=1.0)
xtick_labels, ytick_labels = ax.get_xticklabels(), ax.get_yticklabels()
for label in xtick_labels + ytick_labels:
    label.set_fontsize(24)

ax.tick_params(
    axis="x", which="major", direction="out", length=10, width=2, labelsize=18
)
ax.tick_params(
    axis="x", which="minor", direction="in", length=5, width=2, labelsize=18
)
ax.tick_params(
    axis="y", which="major", direction="out", length=10, width=2, labelsize=18
)
ax.tick_params(
    axis="y", which="minor", direction="in", length=5, width=2, labelsize=18
)

for spine in ax.spines.values():
    spine.set_linewidth(2)
plt.show()
```

```python
with open('renewal.counts.txt', 'r') as file:
    lines = file.readlines()
values = []
for x in lines[:]:
    x = x.strip()
    x = x.split()
    for x in x:
        # print(float(x))
        values.append(float(x))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
ax.bar(range(1, len(values)+1), values, edgecolor='black', linewidth=1.5)
ax.set_xticks(range(1, len(values)+1), [str(i) for i in range(1, len(values)+1)])
xtick_labels, ytick_labels = ax.get_xticklabels(), ax.get_yticklabels()
for label in xtick_labels + ytick_labels:
    label.set_fontsize(24)

ax.tick_params(
    axis="x", which="major", direction="out", length=10, width=2, labelsize=18
)
ax.tick_params(
    axis="x", which="minor", direction="in", length=5, width=2, labelsize=18
)
ax.tick_params(
    axis="y", which="major", direction="out", length=10, width=2, labelsize=18
)
ax.tick_params(
    axis="y", which="minor", direction="in", length=5, width=2, labelsize=18
)

for spine in ax.spines.values():
    spine.set_linewidth(2)
plt.show()
```
